```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency
import numpy as np
import os

# Fungsi untuk mencari dan memuat file Excel
def load_excel_file():
    # Daftar kemungkinan nama dan lokasi file
    possible_files = [
        "Soal UAS Riset Pasar.xlsx",
        "Soal UAS Riset Pasar (1).xlsx",
        "./Soal UAS Riset Pasar.xlsx",
        "./Soal UAS Riset Pasar (1).xlsx",
        "/content/Soal UAS Riset Pasar.xlsx",  # Google Colab
        "/content/Soal UAS Riset Pasar (1).xlsx",
        "/mnt/data/Soal UAS Riset Pasar.xlsx",
        "/mnt/data/Soal UAS Riset Pasar (1).xlsx"
    ]

    print("Mencari file Excel...")
    print("File yang tersedia di direktori saat ini:")
    current_files = []
    try:
        current_files = os.listdir('.')
        excel_files = [f for f in current_files if f.endswith('.xlsx') or f.endswith('.xls')]
        if excel_files:
            for f in excel_files:
                print(f"  - {f}")
        else:
            print("  Tidak ada file Excel (.xlsx/.xls) ditemukan")
    except:
        print("  Tidak dapat mengakses direktori")

    # Coba buka file dari daftar kemungkinan
    for file_path in possible_files:
        if os.path.exists(file_path):
            print(f"\nFile ditemukan: {file_path}")
            try:
                xls = pd.ExcelFile(file_path)
                return xls, file_path
            except Exception as e:
                print(f"Error membaca file {file_path}: {e}")
                continue

    # Jika tidak ditemukan, coba gunakan file Excel pertama yang ditemukan di direktori
    if excel_files:
        file_path = excel_files[0]
        print(f"\nMencoba menggunakan file: {file_path}")
        try:
            xls = pd.ExcelFile(file_path)
            return xls, file_path
        except Exception as e:
            print(f"Error membaca file {file_path}: {e}")

    return None, None

# Muat file Excel
xls, file_path = load_excel_file()

if xls is None:
    print("\n" + "="*50)
    print("FILE EXCEL TIDAK DITEMUKAN!")
    print("="*50)
    print("\nCara mengatasi:")
    print("1. Upload file Excel ke environment ini")
    print("2. Pastikan nama file sesuai: 'Soal UAS Riset Pasar.xlsx' atau 'Soal UAS Riset Pasar (1).xlsx'")
    print("3. Atau letakkan file di direktori yang sama dengan notebook ini")
    print("\nJika Anda menggunakan Google Colab:")
    print("   - Klik ikon folder di panel kiri")
    print("   - Upload file Excel Anda")
    print("   - Jalankan kembali kode ini")

    # Buat data dummy untuk testing
    print("\n" + "="*50)
    print("MEMBUAT DATA DUMMY UNTUK TESTING")
    print("="*50)

    np.random.seed(42)
    n_samples = 200

    tingkat_studi = np.random.choice(['S1', 'S2', 'S3', 'D3', 'SMA'], n_samples,
                                   p=[0.4, 0.2, 0.1, 0.2, 0.1])
    kepedulian_lingkungan = np.random.choice(['Ya', 'Tidak', 'Netral'], n_samples,
                                           p=[0.6, 0.2, 0.2])

    df = pd.DataFrame({
        'Tingkat Studi': tingkat_studi,
        'Apakah Anda peduli dengan isu ramah lingkungan?': kepedulian_lingkungan
    })

    print("Data dummy berhasil dibuat!")
    print(f"Jumlah sampel: {len(df)}")

else:
    print(f"\nBerhasil memuat file: {file_path}")

    # Tampilkan informasi sheet
    print(f"Sheet yang tersedia: {xls.sheet_names}")

    # Coba parse sheet 'Data Responden'
    try:
        df = xls.parse('Data Responden')
        print("Berhasil memuat sheet 'Data Responden'")
    except:
        # Jika gagal, coba sheet pertama
        sheet_name = xls.sheet_names[0]
        print(f"Sheet 'Data Responden' tidak ditemukan, menggunakan sheet: '{sheet_name}'")
        df = xls.parse(sheet_name)

# Normalisasi nama kolom
df.columns = df.columns.str.strip()

print(f"\nInformasi Dataset:")
print(f"Jumlah baris: {len(df)}")
print(f"Jumlah kolom: {len(df.columns)}")
print(f"\nNama kolom:")
for i, col in enumerate(df.columns, 1):
    print(f"{i}. '{col}'")

print(f"\nSample data (5 baris pertama):")
print(df.head())

# Cari kolom yang diperlukan
tingkat_studi_cols = [col for col in df.columns if 'tingkat' in col.lower() and 'studi' in col.lower()]
if not tingkat_studi_cols:
    tingkat_studi_cols = [col for col in df.columns if 'pendidikan' in col.lower() or 'education' in col.lower()]

lingkungan_cols = [col for col in df.columns if 'lingkungan' in col.lower() or 'ramah' in col.lower() or 'environment' in col.lower()]

print(f"\nKolom Tingkat Studi yang ditemukan: {tingkat_studi_cols}")
print(f"Kolom Kepedulian Lingkungan yang ditemukan: {lingkungan_cols}")

# Gunakan kolom yang ditemukan atau kolom default
tingkat_col = tingkat_studi_cols[0] if tingkat_studi_cols else 'Tingkat Studi'
lingkungan_col = lingkungan_cols[0] if lingkungan_cols else 'Apakah Anda peduli dengan isu ramah lingkungan?'

print(f"\nMenggunakan kolom:")
print(f"Tingkat Studi: '{tingkat_col}'")
print(f"Kepedulian Lingkungan: '{lingkungan_col}'")

# Cek apakah kolom ada di dataframe
if tingkat_col in df.columns and lingkungan_col in df.columns:
    # Tampilkan nilai unique
    print(f"\nNilai unique '{tingkat_col}':")
    print(df[tingkat_col].value_counts())

    print(f"\nNilai unique '{lingkungan_col}':")
    print(df[lingkungan_col].value_counts())

    # Hapus nilai NaN jika ada
    df_clean = df[[tingkat_col, lingkungan_col]].dropna()
    print(f"\nJumlah data setelah menghapus NaN: {len(df_clean)}")

    # Buat tabel kontingensi
    tabel_lingkungan = pd.crosstab(df_clean[tingkat_col], df_clean[lingkungan_col])

    print(f"\nTabel Kontingensi:")
    print(tabel_lingkungan)

    # Visualisasi 1: Heatmap
    plt.figure(figsize=(10, 6))
    sns.heatmap(tabel_lingkungan, annot=True, cmap="YlGn", fmt='d',
                cbar_kws={'label': 'Jumlah Responden'})
    plt.title("Kepedulian terhadap Isu Ramah Lingkungan berdasarkan Tingkat Studi",
              fontsize=14, pad=20)
    plt.ylabel("Tingkat Studi", fontsize=12)
    plt.xlabel("Kepedulian terhadap Isu Ramah Lingkungan", fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

    # Visualisasi 2: Bar plot persentase
    plt.figure(figsize=(12, 6))
    tabel_pct = pd.crosstab(df_clean[tingkat_col], df_clean[lingkungan_col],
                           normalize='index') * 100
    tabel_pct.plot(kind='bar', ax=plt.gca(), width=0.8)
    plt.title("Persentase Kepedulian terhadap Isu Ramah Lingkungan berdasarkan Tingkat Studi",
              fontsize=14)
    plt.ylabel("Persentase (%)", fontsize=12)
    plt.xlabel("Tingkat Studi", fontsize=12)
    plt.legend(title="Kepedulian Lingkungan", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    # Uji Chi-Square
    try:
        chi2_env, p_env, dof, expected = chi2_contingency(tabel_lingkungan)

        print(f"\n" + "="*50)
        print("HASIL UJI CHI-SQUARE")
        print("="*50)
        print(f"Chi-square statistic: {chi2_env:.4f}")
        print(f"P-value: {p_env:.6f}")
        print(f"Degrees of freedom: {dof}")

        # Interpretasi hasil
        alpha = 0.05
        print(f"\nInterpretasi ( = {alpha}):")
        if p_env < alpha:
            print(f" P-value ({p_env:.6f}) <  ({alpha})")
            print(" H0 DITOLAK: Ada hubungan yang signifikan antara tingkat studi dan kepedulian terhadap isu ramah lingkungan")
        else:
            print(f" P-value ({p_env:.6f}) >=  ({alpha})")
            print(" H0 DITERIMA: Tidak ada hubungan yang signifikan antara tingkat studi dan kepedulian terhadap isu ramah lingkungan")

        # Expected frequencies
        print(f"\nExpected Frequencies:")
        expected_df = pd.DataFrame(expected,
                                 index=tabel_lingkungan.index,
                                 columns=tabel_lingkungan.columns)
        print(expected_df.round(2))

        # Cramr's V (effect size)
        n = tabel_lingkungan.sum().sum()
        cramers_v = np.sqrt(chi2_env / (n * (min(tabel_lingkungan.shape) - 1)))
        print(f"\nCramr's V (effect size): {cramers_v:.4f}")

        if cramers_v < 0.1:
            effect_interpretation = "sangat lemah"
        elif cramers_v < 0.3:
            effect_interpretation = "lemah"
        elif cramers_v < 0.5:
            effect_interpretation = "sedang"
        else:
            effect_interpretation = "kuat"

        print(f"Interpretasi kekuatan hubungan: {effect_interpretation}")

        print(f"\n" + "="*50)
        print("RINGKASAN HASIL")
        print("="*50)
        print(f" Total responden yang dianalisis: {n}")
        print(f" Chi-square: {chi2_env:.4f}")
        print(f" P-value: {p_env:.6f}")
        print(f" Signifikansi: {'Ya' if p_env < alpha else 'Tidak'} ( = {alpha})")
        print(f" Kekuatan hubungan: {effect_interpretation} (Cramr's V = {cramers_v:.4f})")

    except Exception as e:
        print(f"Error dalam uji chi-square: {e}")

else:
    print(f"\n TIDAK DAPAT MELAKUKAN ANALISIS")
    print(f"Kolom yang diperlukan tidak ditemukan:")
    print(f"  - '{tingkat_col}': {'' if tingkat_col in df.columns else ''}")
    print(f"  - '{lingkungan_col}': {'' if lingkungan_col in df.columns else ''}")

    print(f"\nSilakan:")
    print(f"1. Periksa nama kolom yang tersedia di atas")
    print(f"2. Atau sesuaikan nama variabel tingkat_col dan lingkungan_col")
```

# **5. Apakah preferensi ketertarikan membeli produk buatan mahasiswa berbeda antara mahasiswa laki-laki dan perempuan?**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency
import numpy as np

# Pastikan DataFrame sudah dimuat (dari kode sebelumnya)
# Jika belum, gunakan kode loading yang sudah dibuat sebelumnya

# Cari kolom yang berkaitan dengan jenis kelamin dan ketertarikan produk mahasiswa
gender_cols = [col for col in df.columns if 'jenis' in col.lower() and 'kelamin' in col.lower()]
if not gender_cols:
    gender_cols = [col for col in df.columns if 'gender' in col.lower()]
if not gender_cols:
    gender_cols = [col for col in df.columns if 'kelamin' in col.lower()]

produk_mahasiswa_cols = [col for col in df.columns if 'tertarik' in col.lower() and 'mahasiswa' in col.lower()]
if not produk_mahasiswa_cols:
    produk_mahasiswa_cols = [col for col in df.columns if 'produk' in col.lower() and 'mahasiswa' in col.lower()]

print("Kolom yang berkaitan dengan jenis kelamin:")
for col in gender_cols:
    print(f"  - '{col}'")

print("\nKolom yang berkaitan dengan ketertarikan produk mahasiswa:")
for col in produk_mahasiswa_cols:
    print(f"  - '{col}'")

# Tentukan kolom yang akan digunakan
gender_col = "Jenis Kelamin"
produk_col = "Apakah Anda tertarik membeli produk buatan mahasiswa?"

if gender_col not in df.columns and gender_cols:
    gender_col = gender_cols[0]
    print(f"\nMenggunakan kolom jenis kelamin: '{gender_col}'")

if produk_col not in df.columns and produk_mahasiswa_cols:
    produk_col = produk_mahasiswa_cols[0]
    print(f"Menggunakan kolom ketertarikan produk: '{produk_col}'")

# Cek apakah kedua kolom ada
if gender_col in df.columns and produk_col in df.columns:
    # Bersihkan data dari nilai NaN
    df_clean = df[[gender_col, produk_col]].dropna()

    print(f"\nInformasi Data:")
    print(f"Total responden setelah cleaning: {len(df_clean)}")
    print(f"Data yang dihapus (NaN): {len(df) - len(df_clean)}")

    # Tampilkan distribusi masing-masing variabel
    print(f"\nDistribusi Jenis Kelamin:")
    print(df_clean[gender_col].value_counts())

    print(f"\nDistribusi Ketertarikan Produk Mahasiswa:")
    print(df_clean[produk_col].value_counts())

    # Buat tabel kontingensi antara jenis kelamin dan ketertarikan membeli produk buatan mahasiswa
    tabel_gender_produk = pd.crosstab(df_clean[gender_col], df_clean[produk_col])

    print(f"\nTabel Kontingensi:")
    print(tabel_gender_produk)

    # Hitung persentase untuk analisis lebih mendalam
    tabel_pct_baris = pd.crosstab(df_clean[gender_col], df_clean[produk_col], normalize='index') * 100
    tabel_pct_kolom = pd.crosstab(df_clean[gender_col], df_clean[produk_col], normalize='columns') * 100

    print(f"\nPersentase berdasarkan Jenis Kelamin (baris):")
    print(tabel_pct_baris.round(1))

    print(f"\nPersentase berdasarkan Ketertarikan (kolom):")
    print(tabel_pct_kolom.round(1))

    # Visualisasi 1: Heatmap
    plt.figure(figsize=(8, 6))
    sns.heatmap(tabel_gender_produk, annot=True, cmap="coolwarm", fmt='d',
                cbar_kws={'label': 'Jumlah Responden'})
    plt.title("Ketertarikan Membeli Produk Buatan Mahasiswa berdasarkan Jenis Kelamin",
              fontsize=14, pad=20)
    plt.ylabel("Jenis Kelamin", fontsize=12)
    plt.xlabel("Ketertarikan", fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

    # Visualisasi 2: Stacked Bar Chart
    plt.figure(figsize=(10, 6))
    tabel_gender_produk.plot(kind='bar', stacked=True, ax=plt.gca(),
                            color=['#ff7f7f', '#7fbf7f', '#7f7fff'])
    plt.title("Distribusi Ketertarikan Produk Mahasiswa berdasarkan Jenis Kelamin",
              fontsize=14, pad=20)
    plt.xlabel("Jenis Kelamin", fontsize=12)
    plt.ylabel("Jumlah Responden", fontsize=12)
    plt.legend(title="Ketertarikan", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    # Visualisasi 3: Grouped Bar Chart
    plt.figure(figsize=(10, 6))
    tabel_gender_produk.plot(kind='bar', ax=plt.gca(), width=0.8)
    plt.title("Perbandingan Ketertarikan Produk Mahasiswa berdasarkan Jenis Kelamin",
              fontsize=14, pad=20)
    plt.xlabel("Jenis Kelamin", fontsize=12)
    plt.ylabel("Jumlah Responden", fontsize=12)
    plt.legend(title="Ketertarikan", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    # Visualisasi 4: Percentage Stacked Bar Chart
    plt.figure(figsize=(10, 6))
    tabel_pct_baris.plot(kind='bar', stacked=True, ax=plt.gca(),
                        color=['#ff7f7f', '#7fbf7f', '#7f7fff'])
    plt.title("Persentase Ketertarikan Produk Mahasiswa berdasarkan Jenis Kelamin",
              fontsize=14, pad=20)
    plt.xlabel("Jenis Kelamin", fontsize=12)
    plt.ylabel("Persentase (%)", fontsize=12)
    plt.legend(title="Ketertarikan", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xticks(rotation=45, ha='right')

    # Tambahkan persentase pada bar
    for container in plt.gca().containers:
        plt.gca().bar_label(container, labels=[f'{v:.1f}%' if v > 5 else '' for v in container.datavalues],
                           label_type='center', fontsize=10, fontweight='bold')

    plt.tight_layout()
    plt.show()

    # Uji Chi-Square
    try:
        chi2_gender_produk, p_gender_produk, dof, expected = chi2_contingency(tabel_gender_produk)

        print(f"\n" + "="*60)
        print("HASIL UJI CHI-SQUARE")
        print("="*60)
        print(f"Chi-square statistic: {chi2_gender_produk:.4f}")
        print(f"P-value: {p_gender_produk:.6f}")
        print(f"Degrees of freedom: {dof}")

        # Interpretasi hasil
        alpha = 0.05
        print(f"\nInterpretasi ( = {alpha}):")
        if p_gender_produk < alpha:
            print(f" P-value ({p_gender_produk:.6f}) <  ({alpha})")
            print(" H0 DITOLAK: Ada hubungan yang signifikan antara jenis kelamin dan ketertarikan membeli produk buatan mahasiswa")
        else:
            print(f" P-value ({p_gender_produk:.6f}) >=  ({alpha})")
            print(" H0 DITERIMA: Tidak ada hubungan yang signifikan antara jenis kelamin dan ketertarikan membeli produk buatan mahasiswa")

        # Expected frequencies
        print(f"\nExpected Frequencies:")
        expected_df = pd.DataFrame(expected,
                                 index=tabel_gender_produk.index,
                                 columns=tabel_gender_produk.columns)
        print(expected_df.round(2))

        # Cek asumsi chi-square (expected frequency >= 5)
        min_expected = expected.min()
        print(f"\nAsumsi Chi-Square:")
        print(f"Expected frequency terkecil: {min_expected:.2f}")
        if min_expected >= 5:
            print(" Asumsi terpenuhi (semua expected frequency >= 5)")
        else:
            print("  Asumsi tidak terpenuhi (ada expected frequency < 5)")
            print("   Hasil uji mungkin tidak reliable")

        # Cramr's V (effect size)
        n = tabel_gender_produk.sum().sum()
        cramers_v = np.sqrt(chi2_gender_produk / (n * (min(tabel_gender_produk.shape) - 1)))
        print(f"\nCramr's V (effect size): {cramers_v:.4f}")

        if cramers_v < 0.1:
            effect_interpretation = "sangat lemah"
        elif cramers_v < 0.3:
            effect_interpretation = "lemah"
        elif cramers_v < 0.5:
            effect_interpretation = "sedang"
        else:
            effect_interpretation = "kuat"

        print(f"Interpretasi kekuatan hubungan: {effect_interpretation}")

        # Analisis pola dalam data
        print(f"\n" + "="*60)
        print("ANALISIS POLA DATA")
        print("="*60)

        for gender in tabel_gender_produk.index:
            total_gender = tabel_gender_produk.loc[gender].sum()
            print(f"\n{gender} (n={total_gender}):")
            for ketertarikan in tabel_gender_produk.columns:
                count = tabel_gender_produk.loc[gender, ketertarikan]
                pct = (count / total_gender) * 100
                print(f"   {ketertarikan}: {count} ({pct:.1f}%)")

        # Temukan pola menarik
        print(f"\n INSIGHTS:")

        # Gender mana yang paling tertarik
        tertarik_cols = [col for col in tabel_gender_produk.columns if 'ya' in col.lower() or 'tertarik' in col.lower()]
        if tertarik_cols:
            tertarik_col = tertarik_cols[0]
            max_tertarik_gender = tabel_pct_baris[tertarik_col].idxmax()
            max_tertarik_pct = tabel_pct_baris.loc[max_tertarik_gender, tertarik_col]
            print(f" {max_tertarik_gender} menunjukkan ketertarikan tertinggi ({max_tertarik_pct:.1f}%)")

        # Perbedaan signifikan
        if p_gender_produk < alpha:
            print(f" Terdapat perbedaan signifikan dalam ketertarikan antara jenis kelamin")
        else:
            print(f" Tidak terdapat perbedaan signifikan dalam ketertarikan antara jenis kelamin")

        print(f"\n" + "="*60)
        print("RINGKASAN HASIL")
        print("="*60)
        print(f" Total responden: {n}")
        print(f" Chi-square: {chi2_gender_produk:.4f}")
        print(f" P-value: {p_gender_produk:.6f}")
        print(f" Signifikansi: {'Ya' if p_gender_produk < alpha else 'Tidak'} ( = {alpha})")
        print(f" Kekuatan hubungan: {effect_interpretation} (Cramr's V = {cramers_v:.4f})")
        print(f" Asumsi chi-square: {'Terpenuhi' if min_expected >= 5 else 'Tidak terpenuhi'}")

    except Exception as e:
        print(f"Error dalam uji chi-square: {e}")

else:
    print(f" TIDAK DAPAT MELAKUKAN ANALISIS")
    print(f"Kolom yang diperlukan tidak ditemukan:")
    print(f"  - '{gender_col}': {'' if gender_col in df.columns else ''}")
    print(f"  - '{produk_col}': {'' if produk_col in df.columns else ''}")

    print(f"\nKolom yang tersedia:")
    for i, col in enumerate(df.columns, 1):
        print(f"{i:2d}. '{col}'")

    print(f"\nSilakan:")
    print(f"1. Periksa nama kolom yang benar")
    print(f"2. Atau sesuaikan variabel gender_col dan produk_col")
```

# **9. Platform belanja online mana yang paling populer di kalangan mahasiswa?**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load file Excel
file_path = "Soal UAS Riset Pasar.xlsx"

try:
    # Baca file Excel
    excel_data = pd.ExcelFile(file_path)

    # Cek apakah sheet 'Data Responden' ada
    if 'Data Responden' in excel_data.sheet_names:
        df = excel_data.parse('Data Responden')
    else:
        print("Sheet 'Data Responden' tidak ditemukan.")
        print("Sheet yang tersedia:", excel_data.sheet_names)
        exit()

    # Nama kolom platform (sesuaikan jika berbeda)
    platform_column = 'Platform mana yang Anda gunakan untuk belanja online?'

    # Cek apakah kolom ada dalam dataframe
    if platform_column not in df.columns:
        print(f"Kolom '{platform_column}' tidak ditemukan.")
        print("Kolom yang tersedia:")
        for i, col in enumerate(df.columns, 1):
            print(f"{i}. {col}")
        exit()

    # Bersihkan data: hapus nilai kosong/NaN
    df_clean = df[platform_column].dropna()

    # Hitung frekuensi platform belanja online
    platform_counts = df_clean.value_counts()

    # Cek apakah ada data
    if platform_counts.empty:
        print("Tidak ada data platform yang valid.")
        exit()

    # Tampilkan platform terpopuler
    most_popular_platform = platform_counts.idxmax()
    most_popular_platform_count = platform_counts.max()

    print("="*60)
    print("ANALISIS PLATFORM BELANJA ONLINE")
    print("="*60)
    print(f"Total responden: {len(df_clean)}")
    print(f"Platform paling populer: {most_popular_platform}")
    print(f"Jumlah pengguna: {most_popular_platform_count} responden")
    print(f"Persentase: {(most_popular_platform_count/len(df_clean)*100):.1f}%")
    print()

    print("Frekuensi penggunaan setiap platform:")
    print("-" * 40)
    for platform, count in platform_counts.items():
        percentage = (count / len(df_clean)) * 100
        print(f"{platform}: {count} responden ({percentage:.1f}%)")

    # Visualisasi
    plt.figure(figsize=(12, 8))

    # Buat bar plot dengan warna yang lebih menarik
    colors = plt.cm.Set3(range(len(platform_counts)))
    bars = platform_counts.plot(kind='bar', color=colors, edgecolor='black', linewidth=0.8)

    # Pengaturan plot
    plt.title('Platform Belanja Online yang Paling Populer di Kalangan Mahasiswa',
              fontsize=16, fontweight='bold', pad=20)
    plt.xlabel('Platform', fontsize=12, fontweight='bold')
    plt.ylabel('Jumlah Pengguna', fontsize=12, fontweight='bold')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Tambahkan nilai di atas setiap bar
    ax = plt.gca()
    for i, (platform, count) in enumerate(platform_counts.items()):
        percentage = (count / len(df_clean)) * 100
        ax.text(i, count + 0.5, f'{count}\n({percentage:.1f}%)',
                ha='center', va='bottom', fontweight='bold', fontsize=10)

    # Sesuaikan layout
    plt.tight_layout()

    # Tampilkan plot
    plt.show()

    # Simpan hasil ke file CSV (opsional)
    result_df = pd.DataFrame({
        'Platform': platform_counts.index,
        'Jumlah_Pengguna': platform_counts.values,
        'Persentase': (platform_counts.values / len(df_clean) * 100).round(1)
    })

    output_file = "hasil_analisis_platform.csv"
    result_df.to_csv(output_file, index=False)
    print(f"\nHasil analisis disimpan ke: {output_file}")

except FileNotFoundError:
    print(f"File '{file_path}' tidak ditemukan.")
    print("Pastikan file Excel berada di direktori yang sama dengan script Python ini.")
except Exception as e:
    print(f"Terjadi error: {str(e)}")
    print("Periksa kembali format file Excel dan nama sheet/kolom.")
```

# **10. Apakah mahasiswa yang memilih unik sebagai alasan mencoba produk baru punya minat berbeda terhadap topik tertentu?**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Path file - sesuaikan dengan lokasi file Anda
file_path = "Soal UAS Riset Pasar.xlsx"  # Ganti dengan path yang sesuai

try:
    # Muat ulang file Excel
    excel_data = pd.ExcelFile(file_path)

    # Cek keberadaan sheet
    if 'Data Responden' not in excel_data.sheet_names:
        print("Sheet 'Data Responden' tidak ditemukan.")
        print("Sheet yang tersedia:", excel_data.sheet_names)
        exit()

    # Baca sheet 'Data Responden'
    df = excel_data.parse('Data Responden')

    # Nama kolom yang akan dianalisis
    produk_baru_col = 'Jika ada produk baru, apa yang membuat Anda tertarik mencoba?'
    topik_minat_col = 'Topik apa yang paling Anda minati saat ini?'

    # Validasi keberadaan kolom
    required_columns = [produk_baru_col, topik_minat_col]
    missing_columns = [col for col in required_columns if col not in df.columns]

    if missing_columns:
        print("Kolom yang tidak ditemukan:")
        for col in missing_columns:
            print(f"- {col}")
        print("\nKolom yang tersedia:")
        for i, col in enumerate(df.columns, 1):
            print(f"{i}. {col}")
        exit()

    # Bersihkan data dan hapus nilai kosong
    df_clean = df.dropna(subset=[produk_baru_col, topik_minat_col])

    print("="*80)
    print("ANALISIS PERBANDINGAN MINAT TOPIK")
    print("="*80)
    print(f"Total responden valid: {len(df_clean)}")

    # Filter data mahasiswa yang memilih "Unik" sebagai alasan mencoba produk baru
    # Normalisasi teks untuk pencocokan yang lebih fleksibel
    df_clean['produk_baru_clean'] = df_clean[produk_baru_col].str.strip().str.lower()
    unik_df = df_clean[df_clean['produk_baru_clean'] == 'unik']

    print(f"Responden yang memilih 'Unik': {len(unik_df)}")

    if len(unik_df) == 0:
        print("Tidak ada responden yang memilih 'Unik'.")
        print("Pilihan yang tersedia:")
        unique_choices = df_clean[produk_baru_col].str.strip().value_counts()
        for choice, count in unique_choices.items():
            print(f"- {choice}: {count} responden")
        exit()

    # Hitung frekuensi minat topik untuk kelompok "Unik"
    topik_counts_unik = unik_df[topik_minat_col].value_counts()

    # Hitung frekuensi minat topik untuk semua responden sebagai pembanding
    topik_counts_all = df_clean[topik_minat_col].value_counts()

    # Tampilkan hasil analisis
    print(f"\nPersentase responden 'Unik': {(len(unik_df)/len(df_clean)*100):.1f}%")
    print("\nTOPIK YANG DIMINATI - KELOMPOK 'UNIK':")
    print("-" * 50)
    for topik, count in topik_counts_unik.items():
        percentage = (count / len(unik_df)) * 100
        print(f"{topik}: {count} responden ({percentage:.1f}%)")

    print(f"\nTOPIK YANG DIMINATI - SEMUA RESPONDEN:")
    print("-" * 50)
    for topik, count in topik_counts_all.items():
        percentage = (count / len(df_clean)) * 100
        print(f"{topik}: {count} responden ({percentage:.1f}%)")

    # Gabungkan data dalam satu DataFrame untuk visualisasi
    comparison_df = pd.DataFrame({
        'Kelompok Unik': topik_counts_unik,
        'Semua Responden': topik_counts_all
    }).fillna(0).astype(int)

    # Hitung persentase untuk analisis lebih mendalam
    comparison_pct_df = pd.DataFrame({
        'Kelompok Unik (%)': (topik_counts_unik / len(unik_df) * 100).round(1),
        'Semua Responden (%)': (topik_counts_all / len(df_clean) * 100).round(1)
    }).fillna(0)

    print(f"\nPERBANDINGAN PERSENTASE:")
    print("-" * 50)
    print(comparison_pct_df)

    # Visualisasi 1: Perbandingan Jumlah Absolut
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))

    # Plot 1: Jumlah absolut
    comparison_df.plot(kind='bar', ax=ax1, color=['#4CAF50', '#FF9800'],
                       edgecolor='black', linewidth=0.8, width=0.8)
    ax1.set_title('Perbandingan Minat Topik\n(Jumlah Responden)',
                  fontsize=14, fontweight='bold', pad=20)
    ax1.set_xlabel('Topik yang Diminati', fontsize=12, fontweight='bold')
    ax1.set_ylabel('Jumlah Responden', fontsize=12, fontweight='bold')
    ax1.tick_params(axis='x', rotation=45)
    ax1.grid(axis='y', linestyle='--', alpha=0.7)
    ax1.legend(['Kelompok Unik', 'Semua Responden'])

    # Tambahkan nilai di atas bar
    for container in ax1.containers:
        ax1.bar_label(container, fontsize=10, fontweight='bold')

    # Plot 2: Persentase
    comparison_pct_df.plot(kind='bar', ax=ax2, color=['#4CAF50', '#FF9800'],
                           edgecolor='black', linewidth=0.8, width=0.8)
    ax2.set_title('Perbandingan Minat Topik\n(Persentase)',
                  fontsize=14, fontweight='bold', pad=20)
    ax2.set_xlabel('Topik yang Diminati', fontsize=12, fontweight='bold')
    ax2.set_ylabel('Persentase (%)', fontsize=12, fontweight='bold')
    ax2.tick_params(axis='x', rotation=45)
    ax2.grid(axis='y', linestyle='--', alpha=0.7)
    ax2.legend(['Kelompok Unik (%)', 'Semua Responden (%)'])

    # Tambahkan nilai persentase di atas bar
    for container in ax2.containers:
        ax2.bar_label(container, fmt='%.1f%%', fontsize=10, fontweight='bold')

    plt.tight_layout()
    plt.show()

    # Visualisasi 3: Heatmap perbandingan
    plt.figure(figsize=(12, 6))

    # Siapkan data untuk heatmap
    heatmap_data = comparison_pct_df.T

    # Buat heatmap
    im = plt.imshow(heatmap_data.values, cmap='YlOrRd', aspect='auto')

    # Set ticks dan labels
    plt.xticks(range(len(heatmap_data.columns)), heatmap_data.columns, rotation=45, ha='right')
    plt.yticks(range(len(heatmap_data.index)), heatmap_data.index)

    # Tambahkan nilai di setiap cell
    for i in range(len(heatmap_data.index)):
        for j in range(len(heatmap_data.columns)):
            plt.text(j, i, f'{heatmap_data.iloc[i, j]:.1f}%',
                    ha='center', va='center', fontweight='bold', color='black')

    plt.colorbar(im, label='Persentase (%)')
    plt.title('Heatmap Perbandingan Minat Topik\n(Kelompok Unik vs Semua Responden)',
              fontsize=14, fontweight='bold', pad=20)
    plt.tight_layout()
    plt.show()

    # Analisis insight
    print("\n" + "="*80)
    print("INSIGHT ANALISIS")
    print("="*80)

    # Topik paling diminati kelompok unik
    top_unik = topik_counts_unik.index[0] if len(topik_counts_unik) > 0 else "Tidak ada"
    top_all = topik_counts_all.index[0]

    print(f"Topik paling diminati kelompok 'Unik': {top_unik}")
    print(f"Topik paling diminati semua responden: {top_all}")

    if top_unik == top_all:
        print(" Kelompok 'Unik' memiliki preferensi topik yang sama dengan mayoritas responden")
    else:
        print(" Kelompok 'Unik' memiliki preferensi topik yang berbeda dari mayoritas responden")

    # Simpan hasil ke Excel
    with pd.ExcelWriter('hasil_analisis_topik.xlsx', engine='openpyxl') as writer:
        comparison_df.to_excel(writer, sheet_name='Perbandingan_Jumlah')
        comparison_pct_df.to_excel(writer, sheet_name='Perbandingan_Persentase')
        topik_counts_unik.to_frame('Jumlah').to_excel(writer, sheet_name='Detail_Kelompok_Unik')
        topik_counts_all.to_frame('Jumlah').to_excel(writer, sheet_name='Detail_Semua_Responden')

    print(f"\nHasil analisis disimpan ke: hasil_analisis_topik.xlsx")

except FileNotFoundError:
    print(f"File '{file_path}' tidak ditemukan.")
    print("Pastikan file Excel berada di direktori yang benar.")
except Exception as e:
    print(f"Terjadi error: {str(e)}")
    print("Periksa kembali format file Excel dan nama sheet/kolom.")
```
